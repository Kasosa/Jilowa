# 1. (3 балла) Выпишите состояние списка [7, 3, 9, 4, 2, 5, 6, 1, 8] после каждой итерации алгоритма сортировки вставками. Сколько сравнений и перестановок было выполнено?

```python
def insertion_sort(some_list):
    comparisons = 0
    permutations = 0
    for i in range(1, len(some_list)):
        key = some_list[i]
        j = i - 1
        while j >= 0 and key < some_list[j]:
            some_list[j + 1] = some_list[j]
            j -= 1
            comparisons += 1
            permutations += 1
        some_list[j + 1] = key
        permutations += 1
        print("итерации", i, ":", some_list, "| сравнений:", comparisons, "| перестановок:", permutations)

x = [7, 3, 9, 4, 2, 5, 6, 1, 8]
insertion_sort(x)
```

# 3. (3 балла) Приведите анализ асимптотической сложности алгоритма сортировки. Напишите значения его временной и пространственной сложностей.

Алгоритм сортировки вставками имеет следующую асимптотическую сложность:

# 1. Временная сложность:
  **- В лучшем случае: \( O(n) \) - когда массив уже отсортирован, и каждый новый элемент вставляется в отсортированную последовательность без необходимости перемещать другие элементы.**
  
  **- В среднем и в худшем случаях: \( O(n^2) \) - когда массив обратно отсортирован или содержит обратно упорядоченные элементы. В этом случае каждый новый элемент должен быть вставлен в начало массива, требуя перестановки всех предыдущих элементов.**

# 2. Пространственная сложность:
   **- \( O(1) \) - дополнительная память, используемая только для временных переменных, не зависит от размера входного массива.**

```python
import time
import random
import matplotlib.pyplot as plt

def random_list(n):
    return [random.randint(0, 1000) for _ in range(n)]

def reverse_sorted_list(n):
    lst = list(range(n))
    lst[-1], lst[n//2] = lst[n//2], lst[-1]  # Swapping middle and last elements
    return lst[::-1]

def equal_parts_list(n):
    lst = []
    for i in range(n):
        lst.append(i // (n//10))
    return lst

def partially_ordered_list(n):
    ordered_part = list(range(int(n*0.95)))
    random_part = [random.randint(0, 1000) for _ in range(n - len(ordered_part))]
    return ordered_part + random_part

def nearby_list(n):
    lst = [random.randint(max(0, i-10), min(n-1, i+10)) for i in range(n)]
    return lst

def run_experiment(func, sizes):
    times = []
    for size in sizes:
        start_time = time.time()
        data = func(size)
        elapsed_time = time.time() - start_time
        times.append(elapsed_time)
    return times

def plot_results(title, sizes, times):
    plt.plot(sizes, times, marker='o')
    plt.title(title)
    plt.xlabel('Input Size (n)')
    plt.ylabel('Execution Time (seconds)')
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    sizes = [100, 500, 1000, 2000, 4000, 6000, 8000, 10000]


    random_times = run_experiment(random_list, sizes)
    reverse_sorted_times = run_experiment(reverse_sorted_list, sizes)
    equal_parts_times = run_experiment(equal_parts_list, sizes)
    partially_ordered_times = run_experiment(partially_ordered_list, sizes)
    nearby_times = run_experiment(nearby_list, sizes)


    plot_results("Random List", sizes, random_times)
    plot_results("Reverse Sorted List", sizes, reverse_sorted_times)
    plot_results("Equal Parts List", sizes, equal_parts_times)
    plot_results("Partially Ordered List", sizes, partially_ordered_times)
    plot_results("Nearby List", sizes, nearby_times)
```

# Для этой функции выполните задание из пункта 4. Есть ли отличие в результатах работы функций insertion_sort и insertion_sort_with_bin_search? Как вы можете это объяснить?

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

def binary_search(arr, key, low, high):
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1
    return low

def insertion_sort_with_bin_search(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = binary_search(arr, key, 0, i - 1)
        arr = arr[:j] + [key] + arr[j:i] + arr[i+1:]
    return arr
```
